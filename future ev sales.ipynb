```
columns_2w = ['Engine type', '2W(2021-22)', '2W(2022-23)', '2W(2023-24)']
data_2w = df1[columns_2w]
data_2w.set_index('Engine type', inplace=True)
data_2w.plot(kind='bar', figsize=(14, 8))
plt.title('Number of 2W Vehicles Registered by Engine Type Over the Years')
plt.xlabel('Engine Type')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.tight_layout()

# Show the plot
plt.show()
```

```
from sklearn.linear_model import LinearRegression
import numpy as np

# Extract the EV sales data
ev_sales = df1.loc[df1['Engine type'] == 'EV', ['2W(2021-22)', '2W(2022-23)', '2W(2023-24)']].values.flatten()
years = np.array([2022, 2023, 2024]).reshape(-1, 1)

# Fit a linear regression model
model = LinearRegression()
model.fit(years, ev_sales)

# Predict EV sales for the upcoming years
future_years = np.array([2025, 2026, 2027]).reshape(-1, 1)
predicted_ev_sales = model.predict(future_years)

# Print the predictions
for year, sales in zip(future_years.flatten(), predicted_ev_sales):
    print(f"Predicted EV sales of Bikes for {year}: {int(sales)}")
```
